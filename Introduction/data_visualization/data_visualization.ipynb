- It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

- There is also a procedural "pylab" interface based on a state machine (like OpenGL), designed to closely resemble that of MATLAB, though its use is discouraged.

```python

plt.plot():
```

- The plot() function is used to draw points (markers) in a diagram.

- By default, the plot() function draws a line from point to point.

- The function takes parameters for specifying points in the diagram.

- Parameter 1 is an array containing the points on the x-axis.

- Parameter 2 is an array containing the points on the y-axis.

- If we need to plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.


````python
import numpy as np
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x))
````

In example above, we use np.linspace(0, 10, 1000) to create 1000 evenly spaced points from 0 to 10, both inclusive and store it in x.

Then we use np.sin(x) to create an array of sine values of x and store it in y.

Then we use plt.plot(x, y) to plot the graph.

If we want to add a title to the graph, we use plt.title("Sine Wave").

To add a label to the x-axis, we use plt.xlabel()

To add a label to the y-axis, we use plt.ylabel()

````python
x = np.linspace(0, 10, 1000)

y = np.sin(x)

plt.xlabel("Time")

plt.ylabel("Amplitude")

plt.title("Sine Wave")

plt.plot(x, y)
````

### Scatter Plot:

- A scatter plot is a type of plot that shows the data as a collection of points.

- The position of a point depends on its two-dimensional value, where each value is a position on either the horizontal or vertical dimension.

- Related plots:

    - Bubble chart
    - Scatter plot matrix
    - 3D scatter plot
    - Marginal histogram
    - Marginal boxplot
    - Marginal violin plot
    - Joint kernel density estimate
    - Rug plot
    - Hexbin plot
    - Clustered points

- Scatter plots are used to observe relationships between variables.

- The dots in a scatter plot not only report the values of individual data points, but also patterns when the data are taken as a whole.

- If the pattern is linear, we may summarize that there is a linear relationship between the variables; if it is not, we may summarize that the variables are related nonlinearly.

````python
x = np.linspace(0, 10, 100)
y = np.cos(x)
plt.scatter(x, y)
````

To add a legend to the graph, we use plt.legend(), and to add a title to the graph, we use plt.title("Cosine Waves").

````python
x = np.linspace(0, 10, 100)
y = np.cos(x)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Cosine Wave")
plt.legend(["Cosine"])
plt.scatter(x,y)
````

#### Changing the color of the graph:

- We can change the color of the graph by passing a third parameter to the plot() function.

- The third parameter is a string representing the color of the line.

- The default color is blue, and the possible values of the third parameter are:

    - b: blue
    - g: green
    - r: red
    - c: cyan
    - m: magenta
    - y: yellow
    - k: black
    - w: white

#### Changing the line style of the graph:

- We can change the line style of the graph by passing a third parameter to the plot() function.

- The third parameter is a string representing the line style.

- The default line style is a solid line, and the possible values of the third parameter are:

    - -: solid line
    - --: dashed line
    - -. : dash-dot line
    - : dotted line



````python
x = np.linspace(0, 10, 100)
y = np.arccos(x)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Inverse Cosine Wave")
plt.legend(["Inverse Cosine"])
plt.plot(x,y, color='green', linestyle='dashed', linewidth = 3, marker='o', markerfacecolor='blue', markersize=12)
````

It's possible to get two or more plots in the same graph by using the plot() function multiple times.

````python
first_x_plot = np.linspace(0, 10, 100)
first_y_plot = np.sin(first_x_plot)
plt.plot(first_x_plot, first_y_plot, label="Sine", color='b')

second_x_plot = np.linspace(0, 10, 100)
second_y_plot = np.cos(second_x_plot)
plt.plot(second_x_plot, second_y_plot, label="Cosine", color='r')

plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Sine and Cosine Waves")
plt.legend(["Sine", "Cosine"])

plt.show()
````


- The set() function is used to set aesthetic parameters in one step.

- It's also used to set the aesthetic style of the plots.

- The set() function accepts parameters like:

    - rc: to set the matplotlib parameters
    - style: to set the style of the plot
    - palette: to set the color palette
    - font: to set the font family
    - font_scale: to set the font size
    - color_codes: to set the color codes



````python
import pandas as pd

sns.set()
X = pd.Series(50*np.random.randn(5000))
Y = pd.Series(200*np.random.randn(5000))
Z = pd.Series(100*np.random.randn(5000)+500)

data = pd.DataFrame({'X':X, 'Y':Y, 'Z':Z})

for col in data.columns:
    plt.hist(data[col], density=True, alpha=0.5)
    sns.kdeplot(data[col], shade=True)
````

    - facet_kws: facet keyword arguments
    - legend: to show legend
    - log_scale: to use log scale

    - kde: to plot a kernel density estimate
    - rug: to add a rug plot
    - stat: statistical transformation to use on the data
    - common_norm: if True, normalize the histogram heights so that the sum of all bars equals 1
    - common_bins: if True, use the same number of bins for all histograms
    - discrete: if True, draw a histogram
    - shrink: amount to shrink the legend by
    - bins: specification of hist bins
    - binwidth: width of each bin
    - multiple: stack and center multiple distributions
    - element: matplotlib element to use
    - fill: if True, fill in the area under univariate density curves
    - linewidth: width of the line
    - cut: when using a univariate variable, only plot observations within this range
    - clip: when using a univariate variable, only plot observations within this range
    - legend_out: if True, the figure size will be extended, and the legend will be drawn outside the plot on the center right

````python
import numpy as np
from bokeh.models import Legend
output_file("Bokeh example.html")
p = figure(x_axis_label='X', y_axis_label='Y', width=800, height=400)
p.title.text = 'Bookeh example'

x = np.linspace(0, 10, 30)
y1 = np.sin(x)
y2 = np.cos(x)

line1 = p.line(x, y1, line_color='blue', line_width=2)
circle = p.circle(x, x, fill_color='green', size=5)
line2 = p.line(x, y2, line_color='red', line_width=2)
triangle = p.triangle(x, x**2, fill_color='black', size=5)

legend = Legend(items=[("sin(x)", [line1]), ("cos(x)", [line2]), ("y=x", [circle]), ('x=x**2', [triangle])], location=(0, 0))

p.add_layout(legend, 'right')

show(p)
````

```python

gridplot():
```

- The gridplot() function is used to arrange Bokeh Plots in grid layout.

- It's a convenience method that can generate a plot layout from a collection of plots.

````python
from bokeh.layouts import gridplot

output_file("Bokeh gridplot example.html")

x = np.linspace(0, 6*np.pi, 100)

y0 = np.sin(x)
y1 = np.cos(x)
y2 = np.sin(x) + np.cos(x) -1

s1 = figure()

s1.title.text = 'Figure 1'
s1.title.text_font_size = '20px'
s1.title.align = 'center'
s1.title.text_color = 'purple'

s1.circle (x, y0, size=10, color='navy', alpha=0.5, legend_label='Sin(x)', line_width=2, line_color='purple' )

s1.legend.title = 'Legend'

s2 = figure()

s2.title.text = 'Figure 2'
s2.title.text_font_size = '20px'
s2.title.align = 'center'
s2.title.text_color = 'orange'

s2.triangle(x, y1, size=10, color='firebrick', alpha=0.5, legend_label='Cos(x)', line_width=2, line_color='orange')

s2.legend.title = 'Legend'

s3 = figure()

s3.title.text = 'Figure 3'
s3.title.text_font_size = '20px'
s3.title.align = 'center'
s3.title.text_color = 'olive'


s3.square(x, y2, size=10, color='olive', alpha=0.5, legend_label='Sin(x) + Cos(x) - 1', line_width=2, line_color='olive')

s3.legend.title = 'Legend'

p = gridplot([[s1, s2, s3]], toolbar_location=None)

show(p)
````
