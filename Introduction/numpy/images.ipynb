# Images

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
im = plt.imread('images/photo.jpg')
```

```python
print(f'The type of the image is {type(im)}\nThe shape of the image is {im.shape}')
```

```python
plt.imshow(im)
```

```python
img = im.copy()

red = img[:,:,0]
```

```python
print(f'The shape of the red channel is {red.shape}')
```

```python
fig, ax = plt.subplots(1,3) # fig is representation of the figure, ax is the axes of the figure, in this case we have 3 axes, one for each channel of the image (RGB)
# The return of subplots is a tuple with the figure and the axes

ch = ['Red', 'Blue', 'Green']

for i, axi in enumerate(ax.flat): # the enumerate is used so we can use two or more variables in a for loop to store informations of the iterable
    # ax is representated by [0][0], [0][1], [0][2] in this case (1,3)
    #The .flat is used to threat the axes as a 1D array instead of a 2D array

    axi.imshow(img[:,:,i], plt.cm.gray) # axi is each subplot in the figure. 
    # The .imshow is used to show the image in the subplot, the parameters are the image and the color map
    # The slice [:,:,i] is used to select the channel of the image, in this case the i is the index of the for loop
    # The plt.cm.gray is used to show the image in gray scale
    # i goes to 0, 1, 2, so the first subplot will be the red channel, the second the green and the third the blue

    axi.set(xticks=[], yticks=[], title=ch[i]) # the .set a function that sets the properties of the axes, in this case we are setting the title of the subplot and removing the ticks
    # The title is the name of the channel, in this case we are using the ch list to set the title
    # The ticks are the numbers in the axes, in this case we are removing them, we are setting the xticks and yticks to an empty list
    
    axi.set_xlabel(ch[i], color=ch[i].lower()) # xlabel is the label of the x axis, in this case we are setting the label to the name of the channel and the color to the name of the 
    # channel in lower case, we are using the .lower() to convert the string to lower case to use as the color it appears below the image in the subplot

    fig.set_size_inches(20,50) # The .set_size_inches is used to set the size of the figure, in this case we are setting the size to 20x50
```
