# Expectations

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Bernoulli

$ E[X] = p $ Expected value of success probability

```python
p = 0.812
sample_mean_X = []

for N in range(1, 10000):
    sample_mean_X.append((np.random.rand(N) <= p).mean())

y = np.array(sample_mean_X)

x = np.arange(y.size)

plt.plot(x, y, linestyle='--')

```

As the size increases the mean of the sample converges to the mean of the distribution.

## Geometric (p)

$ E[X] = \frac{1}{p} $ Expected value of success probability

```python
# E[X] = 1/p  -> Expected value of success
```

```python
z = np.random.geometric(p=0.35)
print(z) # return the number of trials needed to get the first success
```

```python
z = np.random.geometric(p=0.35, size=10000)
print(z) # return the number of trials needed to get the first success
```

```python
p = 0.312

expected_value = 1/p
print('The expected value is: ', expected_value)

sample_mean_X = []

for N in range(1, 10000):
    sample_mean_X.append((np.random.geometric(p,N)).mean()) # N is the size, p is the probability of success

y = np.array(sample_mean_X)

x = np.arange(y.size)

plt.plot(x, y, linestyle='--')
```

## Binomial (n, p)

$ E[X] = np $ Expected value of success probability

```python
n = 350 # number of trials
p = 0.312 # probability of success

expected_value = n*p
print('The expected value is: ', expected_value)

sample_mean_X = []

for N in range(1, 10000):
    sample_mean_X.append((np.random.binomial(n ,p ,N).mean()))

y = np.array(sample_mean_X)

x = np.arange(y.size)

plt.plot(x, y, linestyle='--')
```

## Normal (μ, σ) 

$ E[X] = μ $ Expected value of success probability

```python
mu = 2.4 # mean
sigma = 3.1 # standard deviation

expected_value = mu
print('The expected value is: ', expected_value)

sample_mean_X = []

for N in range(1, 10000):
    sample_mean_X.append((np.random.normal(mu, sigma, N).mean()))

y = np.array(sample_mean_X)

x = np.arange(y.size)

plt.plot(x, y, linestyle='--')
```
