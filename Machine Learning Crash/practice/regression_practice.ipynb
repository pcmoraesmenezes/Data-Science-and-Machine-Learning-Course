# Regression Practice

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
````

````python
n = 20

X = np.arange(n) # 0, 1, 2, ..., n-1
y = 4*X + 3*(X**2) - 100

plt.scatter(X,y)
````

The task of regression is to use the data in `X` to predict the values in `y`.

Although the name of model is called `LinearRegression`, there are ways to extend to non-linear models.

I'm going to try to map the function that generates `y` to the function that generates `X` using a linear model.

````python
model = LinearRegression()
model.fit(X[:, np.newaxis], y)

y_predict = model.predict(X[:, np.newaxis])

plt.scatter(X,y)
plt.plot(X,y_predict, color='red')
````

Let's explain the code above:

```python

model = LinearRegression()
model.fit(X[:, np.newaxis], y)

y_predict = model.predict(X[:, np.newaxis])

plt.scatter(X,y)
plt.plot(X,y_predict, color='red')

```

First we are creating a model object. This model object is an instance of the `LinearRegression` class. This class has a method called `fit` that takes in the `X` and `y` data and fits the model to the data. The `fit` method is going to find the best parameters for the model to fit the data. The `fit` method is going to store these parameters in the `model` object.

The `model.fit` recieves a 2D array for `X` and a 1D array for `y`. The `X` array is 2D because the `fit` method expects a 2D array. The `y` array is 1D because the `fit` method expects a 1D array. We use the `np.newaxis` to add a new axis to the `X` array. This new axis is the second axis. This new axis is the axis that the `fit` method expects.

The `X` before the `np.newaxis` is a 1D array. The `X` after the `np.newaxis` is a 2D array. The `y` is a 1D array.

`y_predict` is the predicted values of `y` using the `X` data and the model that was fit to the data.

`model.predict` recieves a 2D array for `X` and returns a 1D array for `y`.

The `plt.scatter` plots the `X` and `y` data as points. The `plt.plot` plots the `X` and `y_predict` data as a line.

We can see the model in the red line, and the data in the blue points.

The model not learned well.

````python
poly_model = PolynomialFeatures(degree=2, include_bias=True)

X_poly = poly_model.fit_transform(X[:, np.newaxis])

model = LinearRegression()
model.fit(X_poly, y)

y_predict = model.predict(X_poly)

plt.scatter(X,y)
plt.plot(X,y_predict, color='red')
````

Now let's explain this model:

```python

poly_model = PolynomialFeatures(degree=2, include_bias=True)

X_poly = poly_model.fit_transform(X[:, np.newaxis])

model = LinearRegression()
model.fit(X_poly, y)

y_predict = model.predict(X_poly)

plt.scatter(X,y)
plt.plot(X,y_predict, color='red')

```
First we creating a new model called `poly_model`. This model is an instance of the `PolynomialFeatures` class. This class has a method called `fit_transform` that takes in the `X` data and transforms it into a polynomial. The `fit_transform` method is going to find the best polynomial to fit the data. The `fit_transform` method is going to store these parameters in the `poly_model` object.

The values to the `degree` parameter is the degree of the polynomial. The `include_bias` parameter is whether to include the bias term. The bias term is the constant term in the polynomial. Degree means the highest power of the polynomial. For example, if the degree is 2, then the polynomial is `ax^2 + bx + c`. Bias means the constant term in the polynomial. For example, if the bias is True, then the polynomial is `ax^2 + bx + c`. If the bias is False, then the polynomial is `ax^2 + bx`.

Then we use the `poly.fit_transform` method transforms the `X` data into a polynomial. Then we call the `fit` method on the `model` object. This fits the model to the polynomial data. Then we call the `predict` method on the `model` object. This predicts the `y` values using the polynomial data. Then we plot the `X` and `y` data as points. Then we plot the `X` and `y_predict` data as a line.

We can see the model in the red line, and the data in the blue points.

The model learned well.
