## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

# Make a plot with using matplotlib:

#  x: 0 to 20 and generate 1500 values
#  y: cos of x
#  figure size 6x4
#  style 'seaborn-talk'
#  line color = orange
#  line style = dashed
# do not use axis command


# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

fig = plt.figure(figsize=(6,4))

x = np.linspace(0, 20, 1500)
y = np.cos(x)

plt.style.use('seaborn-talk')
plt.plot(x, y, color='orange', linestyle='dashed')
```

```python
# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

# Take the same plot and do the following:

"""
set x limit of 1 to 14
set y limit of -1 to 1.2

set x label to 'x'
set y label to 'cos(x)'

set title to 'Cosine function'

set the color to red
set the line width to 1.5
set the line style to dotted

now reverse the plot at y axis
DO not use axis command

"""

# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
fig = plt.figure(figsize=(6,4))

plt.xlim(1, 14)
plt.ylim(-1, 1.2)

plt.xlabel('X')
plt.ylabel('Cos(x)')

plt.title('Cosine function')

plt.ylim(1.2, -1) # reversed y plot
plt.plot(x, y, color='red', linestyle='dotted',linewidth=1.5,)
```

```python
# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

# Take the same plot and do the following:

"""

Now do the same task using set command

Set x and y labels

"""

fig = plt.figure(figsize=(6,4))
ax = plt.axes()

x = np.linspace(0, 20, 1500)
y = np.cos(x)

plt.style.use('seaborn-talk')

ax.plot(x,y)
ax.set(xlim=(1, 14), ylim=(-1, 1.2), xlabel='X', ylabel='Cos(x)', title='Cosine function')
```

```python
"""

# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Make a Contour plot. Use X and Y points making a meshgrid with linspace function. Generate 50 values for X and 80 values for y from 5-10

For Z take this function: f = X+ (3*Y)*9*X

Use inferno colormap

Set number of intervals to 40

# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
"""

x = np.linspace(5, 10, 50)
y = np.linspace(5, 10, 80)

X,Y = np.meshgrid(x,y)

Z = X+(3*Y)*9*X

plt.contourf(X,Y,Z, 40, cmap='inferno')
plt.colorbar()
```
