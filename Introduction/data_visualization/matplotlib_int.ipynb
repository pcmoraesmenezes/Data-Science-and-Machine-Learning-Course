Scatter plots are a great way to give you a sense of patterns or trends that exist within a dataset. 

In scatter plots, the data points are represented individually with a dot, circle, or other shape.

The function plt.scatter() is used to create a scatter plot.

The parameter s determines the size of the points in the scatter plot.

## Countour Plots

It is a way of representing three-dimensional data in two dimensions, using contours or color-coded regions for the third dimension.

Contour plots (sometimes called Level Plots) are a way to show a three-dimensional surface on a two-dimensional plane.

It graphs two predictor variables X Y on the y-axis and a response variable Z as contours.

These contours are sometimes called z-slices or iso-response values.



````python
fig = plt.figure(figsize=(4,3), dpi=90)
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y) # meshgrid is used to create a grid of points, X and Y are matrices of points in the grid of x and y. Generate pair of points, exemple:
# x = [0, 1], y = [2,3], X = [[0, 1], [0, 1]], Y = [[2, 2], [3, 3]]

Z = np.power(X, 2) + np.sin(Y) # Z is a matrix of points in the grid of X and Y

plt.contour(X, Y, Z, 90, cmap='RdGy') # contourf is used to create a contour plot, 20 is the number of lines, cmap is the color map, the value after Z is the number of lines

plt.colorbar()
````

## Histograms

Histograms are a useful type of statistics plot for engineers.

A histogram is a type of bar plot that shows the frequency or number of values compared to a set of value ranges.

A histogram takes in a series of data and divides the data into a number of bins.

It then plots the frequency data points in each bin.

The bins are usually specified as consecutive, non-overlapping intervals of a variable.

The bins (intervals) must be adjacent, and are often (but are not required to be) of equal size.

```python

plt.hist(x, bins=None, range=None, density=None, weights=None, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, color=None, label=None, stacked=False, normed=None, *, data=None, **kwargs)
# bins → number of bins
# range → lower and upper range of bins
# density → normalize the data
# cumulative → compute Cumulative Distribution Function (CDF)
# histtype → bar, barstacked, step, stepfilled
# align → left, mid, right
# orientation → horizontal, vertical
# rwidth → relative width of bars as a fraction of bin width
# log → scale the axis in log scale
# color → color of the bars
# label → label for the legend
# stacked → if True, multiple data are stacked on top of each other If False multiple data are arranged side by side if histtype is 'bar' or on top of each other if histtype is 'step'
```




````python
from matplotlib.pyplot import hist


plt.figure(figsize=(4,3), dpi=90)

d1 = np.random.normal(0, 0.7, 1000) # d1 is a normal distribution with mean 0 and standard deviation 0.7 with 1000 points
d2 = np.random.normal(-5,2, 1000)
d3 = np.random.normal(5,3,1000)

plt.hist(d1, alpha=0.5, bins=50, histtype='stepfilled', label='Hist 1') # alpha improves the visibility of the graph, bins is the number of bars. histtype is the type of histogram
plt.hist(d2, alpha=0.5, bins=50, histtype='barstacked', label='Hist 2')
plt.hist(d3, alpha=0.5 ,bins=50, histtype='step', label='Hist 3')

plt.legend(loc='best')
````

## Subplots

subplot() function allows you to plot different things in the same figure.

In Matplotlib, the figure (an instance of the class plt.Figure) can be thought of as a single container that contains all the objects representing axes, graphics, text, and labels.

The axes (an instance of the class plt.Axes) is what we see above: a bounding box with ticks and labels, which will eventually contain the plot elements that make up our visualization.

We can create an instance of the class plt.Axes by using the plt.subplots() function.

If we want to have a figure with 2 rows and 1 column (meaning that the two subplots will be displayed on top of each other), we can write:

```python

fig, ax = plt.subplots(2)
```

If we want to have a figure with 1 row and 2 columns (meaning that the two subplots will be displayed side by side), we can write:

```python

fig, ax = plt.subplots(1, 2)
```


````python
fig, ax = plt.subplots(2,3, figsize=(9,7), sharex='col', sharey='row') # subplots is used to create a grid of plots, the first argument is the number of rows, 
#the second is the number of columns. figsize is the size of the figure, sharex and sharey are used to share the x and y axis between the plots

for i in range(2):
    for j in range(3):
        ax[i, j].text(0.5, 0.5, str((i, j)), fontsize=18, ha='center') # the .text is used to write text in the plot, the first argument is the x axis, the second is the y axis,
        # the third is the text, the fourth is the fontsize, ha is the horizontal alignment. The text is written in the middle of the plot because the x and y axis are 0.5 and 0.5 respectively
        # the fontsize is 18 and the horizontal alignment is center because the text is in the middle of the plot and the text is 18 pixels big
````

## 3D Plots



````python
from mpl_toolkits.mplot3d import Axes3D # Axes3D is used to create 3D plots
````

````python
fig = plt.figure(figsize=(6,6), dpi=90)
ax = plt.axes(projection='3d') # projection is used to create a 3D plot

theta = np.linspace(-4*np.pi, 4*np.pi, 100)

Z = np.linspace(-2, 2, 100)

r = Z**2 + 1

X = r*np.sin(theta)

Y = r*np.cos(theta)

ax.plot(X, Y, Z, color='red', linestyle='--', label='3D Plot')

ax.legend()
plt.legend(loc='best')

plt.title('3D Plot')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
````
