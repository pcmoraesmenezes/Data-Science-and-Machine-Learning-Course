## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

# Make a plot with using matplotlib:

#  x: 0 to 20 and generate 1500 values
#  y: cos of x
#  figure size 6x4
#  style 'seaborn-talk'
#  line color = orange
#  line style = dashed
# do not use axis command


# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

fig = plt.figure(figsize=(6,4))

x = np.linspace(0, 20, 1500)
y = np.cos(x)

plt.style.use('seaborn-talk')
plt.plot(x, y, color='orange', linestyle='dashed')
```
