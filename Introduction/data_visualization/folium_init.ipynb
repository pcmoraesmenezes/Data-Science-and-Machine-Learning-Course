````python
import folium
import pandas as pd
import numpy as np
````

````python
df = pd.read_csv('covid_19_till_19_May_2020.csv')
````

````python
df.head()
````

````python
map_ = folium.Map()
df['Date'] = pd.to_datetime(df['Date'])
df2 = df[df['Date'] == max(df['Date'])]
````

````python
df2
````

````python
for i in range(0, len(df2)):
    folium.Circle(
        location=[df2.iloc[i]['Lat'], df2.iloc[i]['Long']],
        color='crimson',
        fill='crimson',
        radius=int(df2.iloc[i]['Confirmed']),
    ).add_to(map_)

map_
````
