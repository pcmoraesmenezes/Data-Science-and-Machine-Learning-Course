## What is scikit-learn?

Scikit-learn is a Python library for machine learning. It is built on top of NumPy, SciPy, and matplotlib. It features various classification, regression, and clustering algorithms, including support vector machines, random forests, gradient boosting, k-means, and DBSCAN, and is designed to interoperate with the Python numerical and scientific libraries NumPy and SciPy.

````python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True) # Create a linear regression model, the fit_intercept=True means that we want to fit a line with a y-intercept

model.fit(x[:, np.newaxis], y) # Fit the model to the data. np.newaxis is used to add a new dimension to x, so that x[:, np.newaxis] is a 2D array with shape (N,1)

#model fit receives two arguments, the first is the x values, and the second is the y values. The first argument must be a 2D array, so we use np.newaxis to add a new dimension
# to x, so that x[:, np.newaxis] is a 2D array with shape (N,1).
#the second argument must be a 1D array, so we don't need to add a new dimension to y.
#model.fits returns the model itself, so we can use it to predict the y values for the data points.
#the function are used for the model to learn the best line that fits the data points.

#After the ajusts of the model, we can use it to predict the y values for the data points.

yfit = model.predict(x[:, np.newaxis]) # Predict the y values for the data points. The slicing is used to add a new dimension to x, so that x[:, np.newaxis] is a 2D array with shape (N,1)
# model.predict has already a fit model, so we don't need to pass the x and y values again, we just need to pass the x values, and it returns the y values predicted by the model.
#model.predict receives one argument, the x values, and it returns the y values predicted by the model.

plt.scatter(x, y, label="Data Points", color='blue') # Plot the data points

# Plot the regression line
plt.plot(x, yfit, label="Linear Regression Line", color='red')

# Add labels and a legend for clarity
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

# Show the plot
plt.show()
````

``` python
from sklearn.datasets._samples_generator import make_blobs
```

This function generates isotropic Gaussian blobs for clustering.

What' s the meaning of isotropic Gaussian blobs? 

Isotropic means that the covariance matrix of the clusters is the identity matrix.

Gaussian means that the points belonging to a cluster are normally distributed around a mean. Example: if the mean of a cluster is (1, 1), the points belonging to the cluster will be distributed with a higher density around the point (1, 1) and with a lower density the further away you move from the point (1, 1).

Blobs means that the clusters are equal in size.

A cluster is a group of data points that are grouped together due to similarities in their features.

The function returns a tuple of two arrays:

- The first array returned by the function contains the generated samples. The shape of the array is ``(n_samples, n_features)``. The argument ``n_samples`` is the number of samples. The argument ``n_features`` is the number of features.

- The second array returned by the function contains the integer labels for cluster membership of each sample. The shape of the array is ``(n_samples,)``. The argument ``n_samples`` is the number of samples.

Is useful to generate classification datasets of controlled size and complexity.

The function ``make_blobs`` is used to generate classification data



````python
from sklearn.datasets._samples_generator import make_blobs

X, y = make_blobs(n_samples=500, centers=3, cluster_std=1.2)
# n_samples is the number of data points, 
# centers is the number of clusters, 
# cluster_std is the standard deviation of the clusters
# The function returns two arrays, the first is the data points, and the second is the labels of the data points
# X is recieving the data points, and y is recieving the labels of the data points
# X is an array with two dimensions, and y is an array with one dimension
# We use this function to generate data points to test the K-Means algorithm

plt.scatter(X[:,0], X[:,1], c=y, cmap='autumn') # X is an array with two dimensions so we must print the first dimension and the second dimension, this is happening in the slicing 
# X[:,0] is the first dimension, and X[:,1] is the second dimension
# c = y means that we are going to use the labels to color the data points
# cmap='autumn' means that we are going to use the autumn color map to color the data points
````

```python
from sklearn.model_selection import train_test_split
```

This function splits the data into random train and test subsets.

The function recieves the following arguments:

- ``X``: The first argument is the data you want to split. The shape of the array is ``(n_samples, n_features)``. The argument ``n_samples`` is the number of samples. The argument ``n_features`` is the number of features.

- ``y``: The second argument is the labels you want to split. The shape of the array is ``(n_samples,)``. The argument ``n_samples`` is the number of samples.

The function returns a tuple of four arrays:

- The first array returned by the function contains the training data. The shape of the array is ``(n_samples, n_features)``. The argument ``n_samples`` is the number of samples. The argument ``n_features`` is the number of features.

- The second array returned by the function contains the test data. The shape of the array is ``(n_samples, n_features)``. The argument ``n_samples`` is the number of samples. The argument ``n_features`` is the number of features.

- The third array returned by the function contains the training labels. The shape of the array is ``(n_samples,)``. The argument ``n_samples`` is the number of samples.

- The fourth array returned by the function contains the test labels. The shape of the array is ``(n_samples,)``. The argument ``n_samples`` is the number of samples.

Label is a class assigned to an example. Example: if you want to classify fruits, the label of an example could be apple, orange, or pear.

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33) # Split the data into training and testing sets. The test_size=0.33 means that 33% of the data will be used for 
# testing, and 67% for training

# X_train is the training data points, X_test is the testing data points, y_train is the training labels, and y_test is the testing labels

````

```python 

from sklearn.svm import SVC
```

This class implements the Support Vector Classification (SVC).

The class recieves the following arguments:

- ``C``: The first argument is the penalty parameter C of the error term. It controls the trade off between smooth decision boundary and classifying the training points correctly. The argument ``C`` is a float. The default value is 1.0.

- ``kernel``: The second argument is the kernel type to be used in the algorithm. It must be one of 'linear', 'poly', 'rbf', 'sigmoid', 'precomputed' or a callable. The argument ``kernel`` is a string. The default value is 'rbf'.

- ``gamma``: The third argument is the kernel coefficient for 'rbf', 'poly' and 'sigmoid'. If ``gamma`` is 'auto', then 1/n_features will be used instead. The argument ``gamma`` is a float. The default value is 'auto'.

- ``degree``: The fourth argument is the degree of the polynomial kernel function ('poly'). Ignored by all other kernels. The argument ``degree`` is an int. The default value is 3.

- ``coef0``: The fifth argument is the independent term in kernel function. It is only significant in 'poly' and 'sigmoid'. The argument ``coef0`` is a float. The default value is 0.0.



````python
from sklearn.svm import SVC # Support Vector Classifier (SVC)
````

````python
mdl = SVC(gamma='auto') # Create a SVC model, the gamma auto means that the gamma parameter will be automatically calculated
mdl.fit(X_train, y_train) # Fit the model to the data. The function receives two arguments, the first is the training data points, and the second is the training labels.
# So the model will learn.
````

````python
y_predict = mdl.predict(X_test) # Predict the labels for the testing data points. The function receives one argument, the testing data points, and it returns the predicted labels.
````

````python
acc = 100*np.sum(y_predict == y_test) / y_test.size
#np.sum is used to sum all the elements in the array, so if the array is [1,2,3,4], the sum is 10
# the 100 is used to convert the result to a percentage
# y_predict is an array with the predicted labels, and y_test is an array with the real labels
# y_predict == y_test is an array with True and False values, True if the predicted label is equal to the real label, and False if the predicted label is not equal to the real label
# acc is an array with the percentage of the correct predictions
# y_test.size is the number of elements in the array, so if the array is [1,2,3,4], the size is 4
# So, the acc is the number of correct predictions divided by the number of elements in the array, and multiplied by 100 to convert to a percentage

print(f'The accuracy of the model is: {acc}%')
````

```python

from sklearn.ensemble import RandomForestClassifier

```

This class implements a meta estimator that fits a number of decision tree classifiers on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting.

The class recieves the following arguments:

- ``n_estimators``: The first argument is the number of trees in the forest. The argument ``n_estimators`` is an int. The default value is 10.

- ``criterion``: The second argument is the function to measure the quality of a split. The argument ``criterion`` is a string. The default value is 'gini'.

- ``max_depth``: The third argument is the maximum depth of the tree. The argument ``max_depth`` is an int. The default value is None.

- ``min_samples_split``: The fourth argument is the minimum number of samples required to split an internal node. The argument ``min_samples_split`` is an int. The default value is 2.

- ``min_samples_leaf``: The fifth argument is the minimum number of samples required to be at a leaf node. The argument ``min_samples_leaf`` is an int. The default value is 1.

- ``min_weight_fraction_leaf``: The sixth argument is the minimum weighted fraction of the sum total of weights (of all the input samples) required to be at a leaf node. The argument ``min_weight_fraction_leaf`` is a float. The default value is 0.0.

- ``max_features``: The seventh argument is the number of features to consider when looking for the best split. The argument ``max_features`` is an int. The default value is 'auto'.

- ``max_leaf_nodes``: The eighth argument is the grow trees with ``max_leaf_nodes`` in best-first fashion. The argument ``max_leaf_nodes`` is an int. The default value is None.

- ``min_impurity_decrease``: The ninth argument is a node will be split if this split induces a decrease of the impurity greater than or equal to this value. The argument ``min_impurity_decrease`` is a float. The default value is 0.0.

- ``min_impurity_split``: The tenth argument is threshold for early stopping in tree growth. A node will split if its impurity is above the threshold, otherwise it is a leaf. The argument ``min_impurity_split`` is a float. The default value is None.

- ``bootstrap``: The eleventh argument is whether bootstrap samples are used when building trees. The argument ``bootstrap`` is a bool. The default value is True.

- ``oob_score``: The twelfth argument is whether to use out-of-bag samples to estimate the generalization accuracy. The argument ``oob_score`` is a bool. The default value is False.

- ``n_jobs``: The thirteenth argument is the number of jobs to run in parallel. The argument ``n_jobs`` is an int. The default value is None.

- ``random_state``: The fourteenth argument is the seed used by the random number generator. The argument ``random_state`` is an int. The default value is None.

- ``verbose``: The fifteenth argument is the verbosity level. The argument ``verbose`` is an int. The default value is 0.

- ``warm_start``: The sixteenth argument is whether to reuse the solution of the previous call to fit and add more estimators to the ensemble, otherwise, just fit a whole new forest. The argument ``warm_start`` is a bool. The default value is False.

We use this class to create a random forest classifier.




````python
from sklearn.ensemble import RandomForestClassifier # Random Forest Classifier (RFC)
````

````python
mdl = RandomForestClassifier(n_estimators=20) # Create a RFC model, the n_estimators=20 means that the number of trees in the forest is 20

mdl.fit(X_train, y_train) # Fit the model to the data. The function receives two arguments, the first is the training data points, and the second is the training labels.
````

````python
y_predict = mdl.predict(X_test)
acc = 100*np.sum(y_predict == y_test) / y_test.size
print(f'The accuracy of the model is: {acc}%')
````
