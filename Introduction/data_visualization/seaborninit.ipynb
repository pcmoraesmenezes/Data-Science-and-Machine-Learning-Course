````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

````python
data = {'X': np.random.normal(0, 2, 1000),
        'Y': np.random.normal(3, 5, 1000),
        'Z': np.random.normal(-4, 10, 1000)}

df = pd.DataFrame(data)
````

````python
df.head()
````

````python
sns.set() # set the default seaborn style for our plots
plt.figure(figsize=(4,3))
for col in df.columns:
    plt.hist(df[col], density=True, alpha=0.5, bins=30, label=col)
    sns.kdeplot(df[col], fill=True, ) # kdplot is used to plot the kernel density estimate, it is a non-parametric way to estimate the probability density function of a random variable
    # fill=True fills the area under the curve. If it's false it only shows a line plot.

plt.legend()
plt.show()
````

````python
with sns.axes_style('white'):
    sns.jointplot(x='X', y='Y', data=df, kind='kde', fill=True, color='red') # jointplot is used to plot the relationship between two variables and the associated distributions
````

````python
iris = sns.load_dataset('iris')
sns.pairplot(data=iris, hue='species') # hue is used to color the data points according to the species column
````

````python
sns.relplot(x='sepal_length', y='sepal_width', data=iris, height=3, aspect=3, style='species', hue='species') # relplot is used to plot the relationship between two variables
# and the associated distributions in a single plot
# height and aspect are used to set the size of the plot and the aspect ratio respectively
````

````python
iris.species.unique()
````

````python
sns.relplot(x='sepal_length', y='sepal_width', kind='line', data=iris, hue='species', errorbar=None, markers=True, dashes=False, style='species') # errorbar removes the shadows
# hue is for coloring the data points according to the species column
# markers is used to show the data points
# dashes is used to show the lines
# style is used to differentiate the species
````

````python
tips = sns.load_dataset('tips') # load the tips dataset, dataset about tips given to waiters
````

````python
tips.time.unique()
````

````python
sns.relplot(x='total_bill', y='tip', hue='smoker', style='smoker', data=tips, col='time')
````

````python
fmri = sns.load_dataset('fmri') # load the fmri dataset, dataset about brain activity

fmri.head()
````

````python
len(fmri.subject.unique()) 
````

````python
sns.relplot(x='timepoint', y='signal', hue='event', style='event', col='subject', col_wrap=5, kind='line', data=fmri.query("region == 'frontal'"), errorbar=None)
# x is the timepoint, representing the time in seconds
# y is the signal, representing the signal intensity
# hue is the event, representing the type of stimulus
# style is the event, representing the type of stimulus
# col is the subject, representing the subject number
# col_wrap is used to wrap the columns
# kind is used to plot the line plot
# data is the fmri dataset, query is used to filter the data
# errorbar is used to remove the shadows
````

````python
sns.set(style='ticks', color_codes=True)
sns.catplot(x='day', y='total_bill',data=tips, palette='deep',aspect=1.5, kind='box', hue='sex') # pallete is used to set the color of the plot
# kind is used to plot a kind of plot, in this case a box plot, there are other kinds of plots like violin plot, swarm plot, etc.
````
