# Analyse the data of covid-19 till 19 may 2020

```python
import pandas as pd
from datetime import datetime as dt 
```

```python
df = pd.read_csv('covid_19_till_19_May_2020.csv')
```

```python
df.head()
```

```python
df.drop(['Lat', 'Long'], inplace=True, axis=1)
```

```python
df.head()
```

```python
df.rename(columns={'Province/State': 'Province', 'Country/Region': 'Country'}, inplace=True)
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape[0]
```

```python
idx = df['Province'].isnull()

df['Province'][idx] = "Other"
```
