# Matplotlib 

````python
import matplotlib.pyplot as plt
import numpy as np
````

```python

fig = plt.figure() # create a plot figure
ax = plt.axes() # create axes of a plot
```

fig is responsible for the overall appearance of the plot, and axes is responsible for the plotting region within the figure. 



```python 
plt.style.available # list of available styles
```



````python
plt.style.use('bmh')
fig = plt.figure()
ax = plt.axes()
````

````python
fig = plt.figure()
ax = plt.axes() # ax is == axis object
x = np.linspace(0, 10, 1000) # 1000 points between 0 and 10
y = np.sin(x) # y is the sin of x
ax.plot(x, np.sin(x)) # ax.plot is doing an intersection with the axes object, x[0] and y[0] are the first points of the line
plt.plot(x, y) # plt.plot is doing an intersection with the axes object, x[0] and y[0] are the first points of the line
# the first argument is the x axis, the second is the y axis, representing the axle of the graph.
````

````python
fig = plt.figure()
ax = plt.axes()

plt.xlabel('Value')
plt.ylabel('Cos Value')
plt.title('Cosine Graph')

x = np.linspace(0, 10, 1000)
y = np.cos(x)
ax.plot(x,y)
````

## Legends and Titles

```python

ax.set(xlim=(0, 10), ylim=(-2, 2), # set the x and y limits of the plot
       xlabel='x', ylabel='sin(x)', # set the x and y labels
       title='A Simple Plot') # set the title of the plot
```

```python

ax.legend(loc='upper left', frameon=False) # add a legend
```

```python

ax.plot(x, np.sin(x), '-g', label='sin(x)') # add a label to the plot
```

```python

ax.plot(x, np.cos(x), '--b', label='cos(x)') # add a label to the plot
```

To we add a legend, we must use label keyword in the plot function. 



````python
fig = plt.figure()
ax = plt.axes()

N = np.linspace(0, 10, 1000)
x = np.cos(N)
y = np.sin(N)

plt.xlabel('Lenght of space N')
plt.ylabel('Value')

subtitle = ['Cosine Waves', 'Sin Waves'] # subtitle is a list of strings


plt.plot(N, x, label=subtitle[0])
plt.plot(N, y, label=subtitle[1])

plt.title('Cosine and Sine Waves') 
plt.legend(loc='lower left') # loc is the location of the legend
plt.show()
````

````python


from numpy import arccos


fig = plt.figure()
ax = plt.axes()

N = np.linspace(0, 10, 100)
sin_ = np.sin(N)
cos_ = np.cos(N)
arc_sin_ = np.arcsin(N)
arc_cos_ = np.arccos(N)

subtitles = ['Sine Wave', 'Cosine Wave', 'Arc Sin Wave', 'Arc Cos Wave']

plt.xlabel('Lenght of space N')
plt.ylabel('Value')

plt.plot(N, sin_, label=subtitles[0], color='green', linestyle='dotted')
plt.plot(N, cos_, label=subtitles[1], color='red', linestyle='-')
plt.plot(N, arc_sin_, label=subtitles[2], color='orange', linestyle='--')
plt.plot(N, arc_cos_, label=subtitles[3], color='blue', linestyle='-.')


plt.legend(loc='lower left')
plt.title('Sine, Cosine, Arc Cosine And Arc Sin Waves')

plt.show()

````
