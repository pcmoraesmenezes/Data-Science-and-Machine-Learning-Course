# OverFitting

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
n = 20
X = np.arange(n)


y = 3*X-4 + 10*(np.random.randn(X.size))
```

```python
plt.scatter(X,y)
```

```python


polyModel = PolynomialFeatures(degree=10)

X_poly = polyModel.fit_transform(X[:, np.newaxis])

m = LinearRegression()
m.fit(X_poly, y)

y_pred = m.predict(X_poly)
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred, color='red')
```

```python
X2 = X[0:n:5]
y2 = y[0:n:5]



polyModel = PolynomialFeatures(degree=5)

X_poly = polyModel.fit_transform(X2[:, np.newaxis])

m = LinearRegression()
m.fit(X_poly, y2)

y_pred = m.predict(X_poly)
```

```python
plt.scatter(X2,y2)
plt.plot(X2,y_pred, color='red')
```

```python
X_poly_all = polyModel.fit_transform(X[:, np.newaxis])

y_pred = m.predict(X_poly_all)

plt.scatter(X,y)
plt.plot(X,y_pred, color='red')
```

The high is the `ṕolymodel` degree the better the model fits the data. However the high the degree the more likely the model is to overfit the new data.
