## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

# Make a plot with using matplotlib:

#  x: 0 to 20 and generate 1500 values
#  y: cos of x
#  figure size 6x4
#  style 'seaborn-talk'
#  line color = orange
#  line style = dashed
# do not use axis command


# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

fig = plt.figure(figsize=(6,4))

x = np.linspace(0, 20, 1500)
y = np.cos(x)

plt.style.use('seaborn-talk')
plt.plot(x, y, color='orange', linestyle='dashed')
```

```python
# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

# Take the same plot and do the following:

"""
set x limit of 1 to 14
set y limit of -1 to 1.2

set x label to 'x'
set y label to 'cos(x)'

set title to 'Cosine function'

set the color to red
set the line width to 1.5
set the line style to dotted

now reverse the plot at y axis
DO not use axis command

"""

# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
fig = plt.figure(figsize=(6,4))

plt.xlim(1, 14)
plt.ylim(-1, 1.2)

plt.xlabel('X')
plt.ylabel('Cos(x)')

plt.title('Cosine function')

plt.ylim(1.2, -1) # reversed y plot
plt.plot(x, y, color='red', linestyle='dotted',linewidth=1.5,)
```

```python
# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

# Take the same plot and do the following:

"""

Now do the same task using set command

Set x and y labels

"""

fig = plt.figure(figsize=(6,4))
ax = plt.axes()

x = np.linspace(0, 20, 1500)
y = np.cos(x)

plt.style.use('seaborn-talk')

ax.plot(x,y)
ax.set(xlim=(1, 14), ylim=(-1, 1.2), xlabel='X', ylabel='Cos(x)', title='Cosine function')
```

```python
"""

# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Make a Contour plot. Use X and Y points making a meshgrid with linspace function. Generate 50 values for X and 80 values for y from 5-10

For Z take this function: f = X+ (3*Y)*9*X

Use inferno colormap

Set number of intervals to 40

# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
"""

x = np.linspace(5, 10, 50)
y = np.linspace(5, 10, 80)

X,Y = np.meshgrid(x,y)

Z = X+(3*Y)*9*X

plt.contourf(X,Y,Z, 40, cmap='inferno')
plt.colorbar()
```

```python
"""
# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Generate subplots having 2 rows and 4 columns. There should be 8 plots in total. 
Set the color map to coolwarm.

# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
"""



fig, ax = plt.subplots(2, 4, figsize=(4, 3), sharex='col', sharey='row')

for i in range(2):
    for j in range(4):
        ax[i,j].imshow(np.random.rand(255, 255), cmap='coolwarm')
        ax[i,j].set(xticks=[], yticks=[]) # remove ticks and labels

```

```python
"""
# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Make 2 scatter plots

For first plot:

X,Y,Z (0,5, 50) Marker = '.' Color = green

For second plot:

X,Y,Z (10, 9, 70) Marker = '+' Color = yellow

# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
"""

from mpl_toolkits.mplot3d import Axes3D # Axes3D is used to create 3D plots

fig = plt.figure(figsize=(6,4))
ax = plt.axes(projection='3d')

X = np.random.normal(0,5,50)
Y = np.random.normal(0,5,50)
Z = np.random.normal(0,5,50)

ax.scatter(X, Y, Z, marker='.', color='green')

X = np.random.normal(10,9,70)
Y = np.random.normal(10,9,70)
Z = np.random.normal(10,9,70)

ax.scatter(X, Y, Z, marker='+', color='yellow')

```

## Seaborn

```python
""" 

# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Make a relationship plot using iris data set:

-   X = sepal width
-   Y = petal length
-   Use sepal length for hue
-   Use species for style

# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

"""

import seaborn as sns

iris = sns.load_dataset("iris")

sns.relplot(x='sepal_width', y='petal_length', hue='sepal_length', style='species', data=iris,)

```

```python
"""
# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
Make a relationship plot with tips dataset and do the following:

Take size as x and total_bill as y

Divide the columns by smoker category

Set hue and style to the sex

"""

tips = sns.load_dataset("tips")

sns.relplot(x='size', y='total_bill', col='smoker', hue='sex', style='sex' , data=tips)
```
